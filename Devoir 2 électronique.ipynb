```python
import numpy as np

import matplotlib.pyplot as plt
```

# No 1.

```python
def H(w):
    return 1/ (5.455 + 33j *w/500000 + 100500/(1j*w))
```

```python
def G(w):
    return np.sqrt((np.real(H(w)))**2 + (np.imag(H(w)))**2)
    
def P(w):
    return np.arctan(np.imag(H(w))/np.real(H(w)))
```

```python
x = np.linspace(500, 5000000, 1000000)
Phase = P(x)
mag = 20*np.log(G(x))
#mag = G(x)




#plt.style.use('https://raw.githubusercontent.com/dccote/Enseignement/master/SRC/dccote-errorbars.mplstyle')
fig, axs = plt.subplots(2)
#fig.suptitle('Diagramme de Bode')

axs[0].semilogx(x,mag,'k', linestyle='-', marker='None', zorder=0)
axs.flat[0].set(ylabel=('Gain [dB]'))
seuil = x*0 + max(mag)-3
idx = np.argwhere(np.diff(np.sign(mag - seuil))).flatten()
idx
axs[0].plot(x[idx], mag[idx], 'ro', markersize=5)

for i in axs:
    i.tick_params(direction = 'in', which ='both')

axs[1].semilogx(x, Phase,'k-', zorder=0)
axs.flat[1].set(ylabel=('Phase [rad]'))
axs.flat[1].set(xlabel=('Fréquence angulaire [rad]'))
fig.show()
fig.savefig('Devoir 2 électronique-Bode1.pdf')
```

fréquence centrale:

```python
w0 = x[np.argwhere(mag == max(mag)).flatten()]
w0
```

Fréquence de coupure:

```python
wc = x[idx]
wc
```

```python
Q = w0[0] / (wc[1]-wc[0])
Q
```

# No. 2

```python
def H2(w):
    return (6750000 *w)/( 81j*w**2 + 27300000*w - 1175000000000j)
```

```python
def G2(w):
    return np.sqrt((np.real(H2(w)))**2 + (np.imag(H2(w)))**2)
    
def P2(w):
    return np.arctan(np.imag(H2(w))/np.real(H2(w)))
```

```python
x = np.linspace(500, 5000000, 100000)
Phase = P2(x)
mag = 20*np.log(G2(x))
#mag = G(x)




#plt.style.use('https://raw.githubusercontent.com/dccote/Enseignement/master/SRC/dccote-errorbars.mplstyle')
fig, axs = plt.subplots(2)
#fig.suptitle('Diagramme de Bode')

axs[0].semilogx(x,mag,'k', linestyle='-', marker='None', zorder=0)
axs.flat[0].set(ylabel=('Gain [dB]'))
for i in axs:
    i.tick_params(direction = 'in', which ='both')



axs[1].semilogx(x, Phase,'k-', zorder=0)
axs.flat[1].set(ylabel=('Phase [rad]'))
axs.flat[1].set(xlabel=('Fréquence angulaire [rad]'))

fig.show()
fig.savefig('Devoir 2 électronique-Bode2.pdf')
```
